```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
L = [8,10,12,14]
T = [21.481,47.620,118.93,255.87]
```

```python
plt.plot(L, T)
plt.xlabel("L")
plt.ylabel("T")
plt.show()
```

```python
# Scaling of the time it takes to run the Wang-Landau algorithm
# function to fit the data

def time_scaling(L, a, b):
    return a * L**b

popt, pcov = curve_fit(time_scaling, L, T)
print(popt)
```

```python
length = np.linspace(1, 20, 100)
plt.scatter(L, T, marker='o')
plt.plot(length, time_scaling(length, *popt), label=rf"$T = {popt[0]:.3f} + L^{{{popt[1]:.3f}}}$")
plt.xlabel("L")
plt.ylabel("T")
plt.legend()
plt.show() 
```
